```python
fruits = ['apple','cherry','mango','pineapple']
print("Original list:",fruits)

print("second fruit:",fruits[1])
print("last fruit:",fruits[3])

fruits.append('dragon fruit')
print("list after adding:",fruits)

fruits.remove('cherry')
print("list after removing:",fruits)
print("no. of fruits:",len(fruits))

numbers = (10,20,30)
print("original:",numbers)
print("tuple:",numbers[2])

unique_numbers = {2,3,1,4,5}
print("set(no duplicates):",unique_numbers)

unique_numbers.add(6)
print("set after adding:",unique_numbers)

unique_numbers.remove(2)
print("set after removing :", unique_numbers)

set1={1,2,3,4}
set2={3,4,5,6}
print("union:",set1 | set2)
print("intersection:", set1 & set2)
print("difference:",set1 - set2)

student = {
    'name' : 'chikki',
    'age' : 18,
    'course' : 'python',
}
print("dictionary:",student)

print("student age:", student['age'])

student['grade']='A'
print("after adding grade",student)

del student['course']
print("after deleting:",student)

for key, value in student.items():
  print(f"{key}:{value}")

  keys=['id','age','name']
  values=[22,18,'chikki']
  student_info = dict(zip(keys, values))
  print("Student Info from zip:", student_info)
```
