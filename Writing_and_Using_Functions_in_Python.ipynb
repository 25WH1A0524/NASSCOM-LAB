```python
def greet():
    print("Hello! Welcome to Python Functions.")
greet()

def greet_user(name):
    print(f"Hello, {name}!")

greet_user("Maxie")
greet_user("Blair")

def greet_user(name="Guest"):
    print(f"Welcome, {name}!")

greet_user()
greet_user("Ava")

def print_names(*names):
    print("Names passed:", names)
    for name in names:
        print("Hello", name)

print_names("Maxie", "Blair", "Charlie")

def print_info(**info):
    print("User Info:")
    for key, value in info.items():
        print(f"{key} = {value}")

print_info(name="Maxie", age=18, city="Hyd")

def show_details(*args, **kwargs):
    print("Args:", args)
    print("Kwargs:", kwargs)

show_details(1, 2, 3, name="Ava", city="Mumbai")


def multiply(a, b):
    return a * b

def square(n):
    return multiply(n, n)

print("Square of 7:", square(7))

```
